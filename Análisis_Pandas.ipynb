```python
url = 'https://raw.githubusercontent.com/DataMarina4/proyecto-coder/main/Billionaires%20Statistics%20Dataset.csv'
Data_1 = pd.read_csv(url, index_col=0, encoding='UTF-8', sep=',', on_bad_lines='skip')
Data_1
```

```python
# Datos de las columnas
Data_1.columns
```

```python
# Analizamos los datos de Estados Unidos
data_1_usa = Data_1[Data_1['country']=='United States']
data_1_usa
```

```python
# Millonarios self-made menores de 50 con final Worth superior a 5000
data_1_usa_sm_50_5000 = data_1_usa[(data_1_usa['selfMade']==1) &(data_1_usa['age']<50) & (data_1_usa['finalWorth']>5000)]
data_1_usa_sm_50_5000
```
