## **Selección de Datasets**


**Opción 1**: Estadísticas de Billonarios
*   Variables a estudiar: industrias y fuentes hasta demografía y ubicaciones geográficas.
Este dataset puede proporcionar información acerca de la distribución de la riqueza, los sectores comerciales y la demografía de los multimillonarios en todo el mundo.
*   Archivo: Billionaires Statistics Dataset.csv (35 columnas, 2641 filas)
*   Fuente: https://www.kaggle.com/datasets/nelgiriyewithana/billionaires-statistics-dataset



**Opción 2:** Dataset de Salud Mental
*   Variables a estudiar: niveles de estrés, depresión, ansiedad, bienestar subjetivo y uso de servicios de salud mental.
El dataset puede brindar una mejor comprensión de los cambios en la salud mental a nivel mundial durante el período de tiempo especificado. Para rastrear las tendencias en salud mental se tuvo en cuenta diversos factores demográficos; género, situación laboral y región geográfica.
*   Archivo: Mental Health Dataset.csv (17 columnas, 292365 filas)
*   Fuente: https://www.kaggle.com/datasets/divaniazzahra/mental-health-dataset



**Opción 3:** 17k de Juegos de estrategia en la App Store de Apple
*   Variables a estudiar: precios, ventas en aplicación, calificaciones, géneros y fechas de lanzamiento de las versiones.
Con este conjunto de datos se puede obtener información sobre los juegos de estrategia en los celulares, este submercado incluye títulos como Clash of Clans, Plants vs Zombies y Pokemon GO.
Se puede utilizar la cantidad de calificaciones como indicador indirecto del éxito general de un juego y luego determinar que factores hacen que un juego sea exitoso, o se podría medir el estado del mércado a lo largo del tiempo e intentar predecir hacia dónde se dirige (datos de 17007 juegos de estrategia en la App Store de Apple. Fue recogido el 3 de agosto de 2019).
*   Archivo: appstore_games.csv(15 columnas, 17008 filas)
